```python
plt.figure(figsize= (20, 20))

for i in np.arange(1, len(df_only_numbers.columns)):
    plt.subplot(3, 4, i)
    sns.boxplot(df_only_numbers[df_only_numbers.columns[i]])
    plt.title(df_only_numbers.columns[i])
```

## Workout type

```python
df['Workout_Type'].unique()
```

```python
workout_count = df['Workout_Type'].value_counts().values
category = ['Strength', 'Cardio', 'Yoda', 'HIIT']

fig, axes = plt.subplots()
bar = axes.bar(category, workout_count, alpha=.5, edgecolor='black')
axes.bar_label(bar) 
plt.title('Número de alunos por exercício')
plt.show()
```
